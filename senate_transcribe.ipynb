```python
# conda install pytorch==1.13.1 torchvision==0.14.1 torchaudio==0.13.1 pytorch-cuda=11.7 -c pytorch -c nvidia
# pip install --upgrade --no-deps --force-reinstall git+https://github.com/openai/whisper.git

import whisper

# Load the model
model = whisper.load_model("medium",device="cuda")
print(model)
```

```python
import torch # cheack if gpu is enabled
print(torch.cuda.is_available())  # Should return True
```

```python
import time
 
# Start the timer
start_time = time.time()

# Transcribe the audio file
result = model.transcribe("/home/henry/Downloads/APPA_meeting_audio.mp3")

# End the timer
end_time = time.time()

# Print the transcription
print(result["text"])

# Print the elapsed time
elapsed_time = end_time - start_time
print(f"Transcription completed in {elapsed_time:.2f} seconds.")
```

```python
transcription_text = result["text"]

# Split text into sentences
lines = transcription_text.split(". ")  # Split at periods

# Define the output file path
output_file_path = "/home/henry/Downloads/APPA_meeting_transcription.txt"

# Write the transcription to a file, one sentence per line
with open(output_file_path, "w", encoding="utf-8") as file:
    for line in lines:
        file.write(line.strip() + ".\n")  # Add a period and newline

print(f"Transcription saved to: {output_file_path}")
```

```python
#transcription 

# Define the output file path
output_file_path = "/home/henry/Downloads/APPA_meeting_transcription_with_timestamps.txt"

# Write transcription with timestamps to the file
with open(output_file_path, "w", encoding="utf-8") as file:
    for segment in result["segments"]:
        start = segment["start"]
        end = segment["end"]
        text = segment["text"]
        file.write(f"[{start:.2f} - {end:.2f}] {text}\n")

print(f"Transcription with timestamps saved to: {output_file_path}")
```
